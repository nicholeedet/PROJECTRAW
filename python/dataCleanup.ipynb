```python
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
```

```python
#States
ST_file = "Resources/State.csv"
ST_df = pd.read_csv(ST_file)

ST_df = ST_df.set_index('State')

ST_df.head()
```

```python
#population (Alex)
premise_file = "Resources/PopulationNumbers.csv"
pop_df = pd.read_csv(premise_file)
```

```python
pop_df=pop_df.dropna()
```

```python
# Rename the column headers
pop_cols_rename = pop_df.rename(columns={"table with row headers in column A and column headers in rows 3 through 4. (leading dots indicate sub-parts)": "State",
                                         "Unnamed: 11": "2018 Population"})
```

```python
#Create a filtered dataframe from specific columns
pop_cols = ["State", "2018 Population"]
pop_trans= pop_cols_rename[pop_cols].copy()

```

```python
pop_df = pop_trans.drop(pop_df.index[0:5])
   
#pop_df = pop_df.drop(0)

```

```python
pop_df[['drop', 'State']]= pop_df.State.str.split(".",expand=True,)
```

```python
pop_df.drop('drop', axis='columns', inplace=True)
```

```python
pop_df = pop_df.set_index('State')
pop_df.head()
```

```python
#Energy Type (Paul)
EU_file = "Resources/EnergyUsage.csv"
energy_df = pd.read_csv(EU_file)
```

```python
energy_df = energy_df.replace(to_replace="â€”", value="0")
```

```python
energy_df = energy_df.rename(columns={"Other": "Petrolium Other",
                                 "Total":"Total Petrolium"})
```

```python
energy_df.drop(['Total Energy'], axis='columns', inplace=True)
```

```python
energy_df = energy_df.set_index('State')
energy_df.head()
```

```python
#Total Energy Consumed (Ryan)
TEC_file = "Resources/TotalEnergyConsumed.csv"
TEC_df = pd.read_csv(TEC_file)
#list(TEC_df.columns.values)
TEC_df=TEC_df.drop(columns=['Note: Rankings are based on the full source data values.'])
TEC_df=TEC_df.rename(columns={"State": "Code", "Rank": "Consumption Rank"})
TEC_df = TEC_df.set_index('Code')
TEC_df.head()
```

```python
#State Ranking (Kayla)
SR_file = "Resources/SelectedStateRankingsData.csv"
SR_df = pd.read_csv(SR_file)
```

```python
del SR_df["Federal offshore production is not included in the Production Shares."]
```

```python
NewSR_df = SR_df.rename(columns={"State":"Code","Production, U.S. Share": "Production Share",
                                 "Consumption per Capita, Million Btu":"Consumption, Million Btu",
                        "Consumption per Capita, Rank":"Consumption, Rank",
                        "Expenditures per Capita, Dollars":"Expenditures, Dollars",
                          "Expenditures per Capita, Rank":"Expenditures, Rank"})

NewSR_df = NewSR_df.set_index('Code')
NewSR_df.head()
```

```python
final_df = ST_df.merge(energy_df, how='left', 
                                  left_on='State', 
                                  right_on='State')
```

```python
final_df = final_df.merge(pop_df, how='left', 
                                  left_on='State', 
                                  right_on='State')

```

```python
final_df = final_df.merge(TEC_df, how='left', 
                                  left_on='Code', 
                                  right_on='Code')

```

```python
final_df = final_df.merge(NewSR_df, how='left', 
                                  left_on='Code', 
                                  right_on='Code')

```

```python
final_df = final_df.drop(final_df.index[8])
final_df.reset_index()

final_df = final_df.set_index("State")

final_df.head()
```

```python
final_df.to_csv('Final_Data.csv')
```
