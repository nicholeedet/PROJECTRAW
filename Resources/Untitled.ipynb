```python
import csv, json
li = []
with open('AvgStateRate.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    header=next(reader)
    for State,Lat, Lon, Average in reader:
        li.append({
            "State": State,
            "latitude": Lat,
            "longitude": Lon,
            "Average": Average,
           
           "geo": {
                "__type": "GeoPoint",
                "latitude": Lat,
                "longitude": Lon,
            }
        })
with open("avgrate.json", "w") as f:
    json.dump(li, f)
```
